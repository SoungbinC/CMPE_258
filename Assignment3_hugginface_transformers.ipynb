```python
text = """After waiting for well over a month, i finally received them

-In short very versatile shoe, great combination of support, cushion, traction.

-Pros- The best aspect of this shoe is without a doubt the traction, the grooves are thick and deep, features a herringbone traction pattern. Performs well in both dusty and clean courts, compared to my last shoe the AJ36 the traction is for sure an upgrade.

-Cushion/impact protection- The shoe features the full length boom foam, very light and "bouncy" foam. The impact protection is great, i really didnt feel super plush underfoot, but after 3 hour hoop session twice a week for the last 2 weeks my leg havent suffered a major lower body aches or soreness. Again compared to my last shoe the AJ36 the cushion doesnt feel plush like the full length zoom, but nonetheless has great step in comfort and impact protection. I believe 1-5 can play in these unless they are super big or is looking for cushion. The boom foam provides you with both court feel and cushioning.

-Support/lockdown- The lockdown/support is simply elite. On the sides it features TPU sidewalls, combined with the hard heel counter, thicc ankle padding and DOUBLE shank makes this shoe very stable and supportive. And the lacing system is effective and simple, wearing this shoe made me cut and drive faar more confidently compared to my 36, where the uppers didnt lock my feet nearly enough.

-Fit- In my experience compared to reviews online i think this shoe fits tts.

-Material- Simple and effective, mesh,TPU, leather what you find on the market nowadays.

-Conclusion- I really do enjoy this shoe, it doesnt have any major downsides from my experience. In general i prefer the AC12 compared to my AJ36, the better traction+lockdown/support really does give it an edge. I recommend this shoe to most players. """
```

# Text classification

```python
import textwrap

wrapper = textwrap.TextWrapper(width=80, break_long_words=False, break_on_hyphens=False)
print(wrapper.fill(text))
```

```python
from transformers import pipeline

sentiment_pipeline = pipeline('text-classification')
```

```python
sentiment_pipeline(text)
```

# Named entity recognition

```python
ner_pipeline = pipeline('ner')
```

```python
entities = ner_pipeline(text, aggregation_strategy="simple")
print(entities)
```

```python
for entity in entities:
    print(f"{entity['word']}: {entity['entity_group']} ({entity['score']:.2f})")
```

# Question answering

```python
qa_pipeline = pipeline("question-answering")
```

```python
question = "Which shoes is the best?"

outputs = qa_pipeline(question=question, context=text)
outputs
```

# Text summarization

```python
summarization_pipeline = pipeline("summarization")
```

```python
outputs = summarization_pipeline(text, max_length=45, clean_up_tokenization_spaces=True)
print(wrapper.fill(outputs[0]['summary_text']))
```

# Translation

```python
translator = pipeline("translation_en_to_jap", model="Helsinki-NLP/opus-mt-en-jap")
```

```python
from textwrap import wrap

chunks = wrap(text, 400)
translated_chunks = []
for chunk in chunks:
    output = translator(chunk, clean_up_tokenization_spaces=True, min_length=100)
    translated_text = output[0]['translation_text']
    translated_chunks.append(translated_text)

# Combine the translated chunks back into a single string
translated_text = ' '.join(translated_chunks)
print(translated_text)
```

# Zero-shot classification

```python
zero_shot_classifier = pipeline("zero-shot-classification",
                                model="vicgalle/xlm-roberta-large-xnli-anli")
```

```python
text = 'Dieser Tutorial ist großartig! Ich hoffe, dass jemand von Hugging Face meine Universität besuchen wird :)'
classes = ['Treffen', 'Arbeit', 'Digital', 'Reisen']
```

```python
zero_shot_classifier(text, classes, multi_label=True)
```
